```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Example

```python
with plt.xkcd():
    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    
    spines= ax.spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim([-30, 10])

    data = np.ones(100)
    data[70:] -= np.arange(30)

    ax.annotate(
        'THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',
        xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))

    ax.plot(data)

    ax.set_xlabel('time')
    ax.set_ylabel('my overall health')
    fig.text(
        0.5, 0.05,
        '"Stove Ownership" from xkcd by Randall Munroe',
        ha='center')
```

# Plot

```python
proj = 2048
imagesize= 2048**2
data = {'Acquisition'    : proj*imagesize*2,
        'Reduction'      : proj*imagesize*4, 
        'Reconstruction' : proj*imagesize*4, 
        'Filtering'      : proj*imagesize*4,
        'Segmentation'   : proj*imagesize,
        'Labelling'      : proj*imagesize,
        'Measurements'   : 10000,
        'Classification' : 10}


sizes = []
descriptions = []

for item in data.keys():
    descriptions.append(item)
    sizes.append(data[item])


```

```python
with plt.xkcd():
    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    
    spines= ax.spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.plot(sizes)
    ax.set_xticks(range(len(descriptions)))
    ax.set_xticklabels(descriptions,rotation=45,ha="right")
    ax.set_yticks([])
    
    ax.set_ylabel('Amount of data')
    
    ax.set_xlabel('Data evaluation step')
    
    
    ax.set_title('Data amounts per processing step')
```

```python
with plt.xkcd():
    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    
    spines= ax.spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.plot(sizes)
    ax.set_xticks(range(len(descriptions)))
    ax.set_xticklabels(descriptions,rotation=45,ha="right")
    ax.set_yticks([])
    

    ax.set_title('Data amounts per processing step')
    
    ax.annotate(
        '16 to 32 bits',
        xy=(1, sizes[1]), arrowprops=dict(arrowstyle='->'), xytext=(1.1, sizes[1]*0.7))
    
    ax.annotate(
        '32 to 8 bits',
        xy=(4, sizes[4]), arrowprops=dict(arrowstyle='->'), xytext=(4.5, sizes[4]*2))
    
    ax.annotate(
        'Image to list',
        xy=(6, sizes[6]), arrowprops=dict(arrowstyle='->'), xytext=(6.5, sizes[4]))
    
    ax.set_ylabel('Amount of data')
    
    ax.set_xlabel('Data evaluation step')
```
