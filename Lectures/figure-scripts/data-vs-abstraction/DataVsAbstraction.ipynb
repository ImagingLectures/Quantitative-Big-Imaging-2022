```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Plot

```python
proj = 2048
imagesize= 2048**2
data = {'Acquisition'    : [proj*imagesize*2,1,'pixels'],
        'Reduction'      : [proj*imagesize*4,1,'pixels'], 
        'Reconstruction' : [proj*imagesize*4,1,'pixels'], 
        'Filtering'      : [proj*imagesize*4,1,'pixels'],
        'Segmentation'   : [proj*imagesize,10,'pixel regions'],
        'Labelling'      : [proj*imagesize,300,'items'],
        'Measurements'   : [10000,1000,'list'],
        'Classification' : [100,3000,'categories'],
        'Inference'      : [100,6000,'decisions']}


sizes = []
descriptions = []
abstraction = []
information = []

for item in data.keys():
    descriptions.append(item)
    sizes.append(data[item][0])
    abstraction.append(data[item][1])
    information.append(data[item][2])

```

## Quantiative understanding

```python
with plt.xkcd():
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    
    spines= ax.spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    ax.plot(np.array(abstraction),label='Abstraction')
    
    ax.set_xticks(range(len(descriptions)))
    ax.set_xticklabels(descriptions,rotation=45,ha="right")
    ax.set_yticks([])
    
    ax.set_ylabel('Level of abstraction')
    
    ax.set_xlabel('Data evaluation step')
    
    an1 = ax.annotate("Data", xy=(1, 0.1*abstraction[-1]), xycoords="data",
                  va="bottom", ha="center",
                  bbox=dict(boxstyle="round", fc="w"))
    
    an2 = ax.annotate("Knowledge", xy=(len(abstraction)-2.5, 0.9*abstraction[-1]), xycoords="data",
                  va="bottom", ha="center",
                  bbox=dict(boxstyle="round", fc="w"))
        
    ax.set_title('Quantitative understanding')
    
plt.savefig('DataKnowledge.svg')
plt.savefig('DataKnowledge.pdf')
```

## Total amount of data

```python
with plt.xkcd():
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    
    spines= ax.spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    datasum = np.cumsum(np.array(sizes))
    ax.plot(datasum)
    ax.set_xticks(range(len(descriptions)))
    ax.set_xticklabels(descriptions,rotation=45,ha="right")
    ax.set_yticks([])
    
    ax.set_ylabel('Amount of data')
    
    ax.set_xlabel('Data evaluation step')
    
    
    ax.set_title('Data produced during analysis')
    
    an1 = ax.annotate("Pixels", xy=(2, 0.15*datasum[-1]), xycoords="data",
                  va="bottom", ha="center",
                  bbox=dict(boxstyle="round", fc="w"))
    
    an2 = ax.annotate("Regions", xy=(4.5, 0.15*datasum[-1]), xycoords="data",
                  va="bottom", ha="center",
                  bbox=dict(boxstyle="round", fc="w"))
    
    an3 = ax.annotate("Tables", xy=(7, 0.15*datasum[-1]), xycoords="data",
                  va="bottom", ha="center",
                  bbox=dict(boxstyle="round", fc="w"))

plt.savefig('DataProduced.svg')
plt.savefig('DataProduced.pdf')
```

## Data for each evaluation step

```python
with plt.xkcd():
    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    
    spines= ax.spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.plot(sizes)
    ax.set_xticks(range(len(descriptions)))
    ax.set_xticklabels(descriptions,rotation=45,ha="right")
    ax.set_yticks([])
    

    ax.set_title('Data amount per evaluation step')
    
    ax.annotate(
        '16 to 32 bits',
        xy=(1, sizes[1]), arrowprops=dict(arrowstyle='->'), xytext=(1.1, sizes[1]*0.7))
    
    ax.annotate(
        '32 to 8 bits',
        xy=(4, sizes[4]), arrowprops=dict(arrowstyle='->'), xytext=(4.5, sizes[4]*2))
    
    ax.annotate(
        'Image to list',
        xy=(6, sizes[6]), arrowprops=dict(arrowstyle='->'), xytext=(6.5, sizes[4]))
    
    ax.set_ylabel('Amount of data')
    
    ax.set_xlabel('Data evaluation step')
```
